```python
# Mathematical Operations on Series
import pandas as pd
seriesA = pd.Series([1,2,3,4,5], index =['a', 'b', 'c', 'd', 'e'])
seriesA
```

```python
seriesB=pd.Series([10,20,-10,-50,100],index=['z','y','a','c','e'])
seriesB
```

```python
#that the output of addition is NaN if one of the elements or both elements have no value.
seriesA+seriesB
```

```python
#Subtraction of two series
seriesA-seriesB
```

```python
#Multiplicaation of two series
seriesA*seriesB
```

```python
#Devision of two series
seriesA/seriesB
```

```python
# Creation of DataFrame from Dictionary of Series
ResultSheet={
'Arnab': pd.Series([90, 91, 97],index=['Maths','Science','Hindi']),
'Ramit': pd.Series([92, 81, 96],index=['Maths','Science','Hindi']),
'Sahir': pd.Series([89, 91, 88],index=['Maths','Science','Hindi']),
'Riya': pd.Series([81, 71, 67],index=['Maths','Science','Hindi']),
'Mallika': pd.Series([94, 95, 99],index=['Maths','Science','Hindi'])
}
ResultDF = pd.DataFrame(ResultSheet)
ResultDF
```

```python
Result={
'Maths':pd.Series([90,87,80,78,70],index=['Arnab','Ramit','Sahir','Riya','Mallika']),
'Science':pd.Series([87,95,42,54,56],index=['Arnab','Ramit','Sahir','Riya','Mallika']),
'Hindi':pd.Series([98,87,85,96,87],index=['Arnab','Ramit','Sahir','Riya','Mallika'])
}
DFResult=pd.DataFrame(Result)
DFResult
```

```python
#Operations on rows and columns in DataFrames
#Adding a New Column to a DataFrame
DFResult['English']=[92,87,52,65,45]
DFResult
```

```python
#Adding a New Row to a DataFrame
#DataFRame.loc[] method can also be used to change the data values of a row to a particular value.
DFResult.loc['Jimmy']=[85,65,98,78]
DFResult
```

```python
#Deleting Rows or Columns from a DataFrame
#delete the row with label 'Sahir'
DFResult = DFResult.drop('Sahir', axis=0)
DFResult
```

```python
#Delete the columns having labels  'English' and 'Hindi':
DFResult=DFResult.drop(['English','Hindi'],axis=1)
DFResult
```

```python
#Renaming Row Labels of a DataFrame
#To rename the column indices Maths to sub1, Science to sub2, Hindi to sub3 and English to sub4 we can write the following statement:
DFResult=DFResult.rename({'Maths':'Sub1','Science':'Sub2'}, axis=1)
print(DFResult)
```

```python
#Accessing DataFrames Element through Indexing
#Label Based Indexing
DFResult.loc['Arnab']
```

```python
#To get ranged rows
DFResult.Sub1['Arnab':'Riya']
```

```python
#To get multiple rows from DataFrame
DFResult.loc[['Arnab','Riya']]
```
